```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline   
#내부에서 바로 그려지게 해주는 코드
```

5-1-1 1차원 자료

```python
np.random.seed(1000)
y = np.random.standard_normal(20)
```

```python
x = range(len(y))
plt.plot(x,y)
```

```python
plt.plot(y.cumsum()) #자료의 누적 합 y축 :누적합된 값, x축 y개수
```

```python
plt.plot(y.cumsum())
plt.grid(True) #격자(grid)
plt.axis('tight')  #축 간격 더 조밀하게<-? 차이없어보이는데
```

```python
plt.plot(y.cumsum())
plt.grid(True)
plt.xlim(-1,20)   #lim <- 축 범위 변환
plt.ylim(np.min(y.cumsum())-1,np.max(y.cumsum())+1)
```

```python
plt.figure(figsize = (7,4))
#figsize <- 그림의 (폭,높이)정의
plt.plot(y.cumsum(),'b',lw = 1.5) #black -'b'
plt.plot(y.cumsum(),'ro') #red circle -'ro'
plt.grid(True)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A SImple Plot')
```

5-1-2 2차원 자료

```python
np.random.seed(2000)
y = np.random.standard_normal((20,2)).cumsum(axis = 0)
```

```python
plt.figure(figsize = (7,4))
plt.plot(y,lw =1.5) #두개의 선
plt.plot(y,'go') #두개의 점
plt.grid(True)
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot 2')
```

```python
#주석 붙이기
plt.figure(figsize = (7,4))
plt.plot(y[:,0],lw =1.5,label = '1st')
plt.plot(y[:,1],lw =1.5,label = '2nd')   #선 한개씩 따로 지정하며 주석달기
plt.plot(y,'go') #두개의 점
plt.grid(True)
plt.legend(loc = 0)  #라벨 위치 정하는 거
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot 2')
```

```python
fig, ax1 = plt.subplots()
plt.plot(y[:,0],'b',lw=1.5,label='1st')
plt.plot(y[:,0],'ro')
plt.grid(True)
plt.legend(loc = 8)  #라벨 위치 정하는 거
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A Simple Plot 4')
ax2 = ax1.twinx()
plt.plot(y[:,1],'g',lw=1.5,label='2nd')
plt.plot(y[:,1],'ro')
plt.legend(loc = 0)
plt.ylabel('value 2nd')
```

```python
plt.figure(figsize=(7,5))
plt.subplot(211)
plt.plot(y[:,0],lw = 1.5,label = '1st')
plt.plot(y[:,0],'ro')
plt.grid(True)
plt.legend(loc = 0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot 5')
plt.subplot(212)
plt.plot(y[:,1],'g',lw = 1.5,label ='2nd')
plt.plot(y[:,1],'ro')
plt.grid(True)
plt.legend(loc = 0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
```

```python
plt.figure(figsize=(9,4))
plt.subplot(121)
plt.plot(y[:,0],lw = 1.5,label = '1st')
plt.plot(y[:,0],'ro')
plt.grid(True)
plt.legend(loc = 0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('1st data set')
plt.subplot(122)
plt.bar(np.arange(len(y)),y[:,1],width = 0.5,color = 'g', label = '2nd')
plt.grid(True)
plt.legend(loc = 0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('2nd data set')
```

```python
from matplotlib.patches import Polygon
def func(x):
    return 0.5*np.exp(x)+1
a,b = 0.5,1.5
x = np.linspace(0,2)
y = func(x)

fig, ax = plt.subplots(figsize = (7,5))
plt.plot(x,y,'b',linewidth = 2)
plt.ylim(ymin = 0)

Ix = np.linspace(a,b)
Iy = func(Ix)
verts = [(a,0)] + list(zip(Ix,Iy))+[(b,0)]
poly = Polygon(verts,facecolor = '0.7',edgecolor = '0.5')
ax.add_patch(poly)

plt.text(0.5*(a+b),1,r"$\int_a^b f(x)\mathrm{d}x$",
        horizontalalignment = 'center', fontsize = 20)
plt.figtext(0.9,0.075,'$x$')
plt.figtext(0.075,0.9,'$f(x)$')
ax.set_xticks((a,b))
ax.set_xticklabels(('$a$','$b$'))
ax.set_yticks([func(a),func(b)])
ax.set_yticklabels(('$f(a)$','$f(b)$'))
plt.grid(True)
```

5-3 3차원 플롯

```python
strike = np.linspace(50,150,24)
ttm = np.linspace(0.5,2.5,24)
strike, ttm = np.meshgrid(strike,ttm)
```

```python
#2차원 배열로 변환
strike[:2]
```

```python
iv = (strike-100)**2/(100*strike)/ttm
#내제 변동성 생성
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(9,6))
ax = fig.gca(projection='3d')  #rstride,cstride - 행,열간격 cmap 칼라맵
surf = ax.plot_surface(strike,ttm,iv,rstride = 2,cstride = 2,
                      cmap = plt.cm.coolwarm, linewidth = 0.5,
                      antialiased = True)
ax.set_xlabel('strike')
ax.set_ylabel('time_to_maturity')
ax.set_zlabel('implied volatility')

fig.colorbar(surf,shrink=0.5,aspect=5)
```

```python
#스캐터플롯으로 표시 후 보이는 각도 변경
fig = plt.figure(figsize=(8,5))

ax = fig.add_subplot(111,projection = '3d')
ax.view_init(30,60)  #이함수로 각도변경
ax.scatter(strike,ttm,iv,zdir='z',s=25,
          c='b',marker='o')

ax.set_xlabel('strike')
ax.set_ylabel('time_to_maturity')
ax.set_zlabel('implied volatility')
```
